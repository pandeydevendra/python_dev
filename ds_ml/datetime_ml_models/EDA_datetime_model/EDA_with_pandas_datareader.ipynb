```python
import pandas_datareader as pdr
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
%matplotlib inline
```

```python
df_msft = pdr.get_data_yahoo('MSFT')
```

```python
# 3. Which is the date of the highest price of the stock?
```
