#### matplotlib variations:

1. plot a normal graph with an array of continuous data:

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
efficiency = np.array([0.99, 0.95, 0.93, 0.90, 0.88, 0.85, 0.82, 0.79, 0.77, 0.66])
power = np.array(list(range(10)))
```

```python
plt.plot(efficiency, power)
plt.show()
```

#### 2. plot categorical variables/data:

```python
import matplotlib.pyplot as plt
import numpy as np
class_variable = ["dog", "cat", "cow", "fish"]
numbers = [10, 15, 20, 10]
plt.bar(class_variable, numbers)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
class_variable = ["dog", "cat", "cow", "fish"]
numbers = [10, 15, 20, 10]
plt.pie(numbers, labels = class_variable)
plt.show()
```

#### 3. plot graph to find the relation between two variables using scatter plot:

```python
import matplotlib.pyplot as plt
import numpy as np
var_a = np.array([5, 6, 4, 3, 2, 3, 2, 10, 15, 11, 4, 20, 12, 9])
var_b = np.array([99, 88, 95, 86, 111, 89, 102, 84, 76, 77,  79, 85, 69, 76])
plt.scatter(var_a, var_b)
plt.show()
```

conclusion: var_b decreases with increase in var_a

#### 4. formatting style of plot: decoration/beautification:

axis label, legend, title, grid, color, dashed lines, marker

```python
import matplotlib.pyplot as plt
import numpy as np
efficiency = np.array([0.99, 0.95, 0.93, 0.90, 0.88, 0.85, 0.82, 0.79, 0.77, 0.66])
power = np.array(list(range(10)))

plt.plot(efficiency, power, label = "efficiency - power curve", linestyle = "dashed", marker = '*', color = "red")

plt.xlabel("efficiency")
plt.ylabel("power")

plt.grid("on")

plt.legend()

plt.show()
```
