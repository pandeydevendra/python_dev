```python
def convert_temperature_cell(cell):
    if cell == "NaN":
        return 'temp'
    return cell

df = pd.read_csv("weather_data.csv", converters = {
        'temperature': convert_temperature_cell,
        'event': convert_temperature_cell
})
df

# converter is compatible with excel file, not csv file
```
