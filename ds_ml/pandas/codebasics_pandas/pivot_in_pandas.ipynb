```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("weather.csv")
df
```

```python
df.pivot(index='city',columns='date')
```

```python
df.pivot(index='date', columns = 'city')
```

```python
df.pivot(index='city',columns='date', values="humidity")
```

```python
df.pivot(index='city',columns='date', values="temperature")
```

```python
df.pivot(index='date',columns='city', values="temperature")
```

```python
df.pivot(index='date',columns='city', values="humidity")
```

```python
df.pivot(index='humidity',columns='city')
```

```python
df.pivot(index='humidity',columns='date')
```

```python
df.pivot(index='humidity',columns='temperature')
```

```python
df_NaN = df.pivot(index='humidity',columns='temperature')
df_NaN.fillna(0)
```

```python
df_NaN.replace({
    'NaN': 0,
    None: 'no city'
})
```

## <font color='green'> Pivot Table</font>

```python
df = pd.read_csv("weather2.csv")
df
```

```python
df.pivot_table(index="city",columns="date")
```

```python
df.pivot_table(index="date",columns="city")
```

## <font color = 'black'> Margins

```python
df.pivot_table(index="city",columns="date", margins=True,aggfunc=np.sum)
```

```python
df.pivot_table(index="date",columns="city", margins=True,aggfunc=np.sum)
```

## <font color = 'b'>Grouper

```python
df = pd.read_csv("weather3.csv")
df
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.pivot_table(index=pd.Grouper(freq='M',key='date'),columns='city')
```
