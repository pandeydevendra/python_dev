#### converters:

```python
def convert_people_cell(cell):
    if cell == "n.a.":
        return 'sam walton'
    return cell

df1 = pd.read_excel("stock_data.xlsx", "Sheet1", converters = {
        'people': convert_people_cell
})
df1
```

```python
def convert_people_cell(cell):
    if cell == "n.a.":
        return 'sam walton'
    return cell

def convert_eps_cell(cell):
    if cell == "not available":
        return None
    return cell

df1 = pd.read_excel("stock_data.xlsx", "Sheet1", converters = {
    'people': convert_people_cell,
    'eps' : convert_eps_cell
})
df1
```

### Merging two DataFrames (excel files:

```python
df_stocks = pd.DataFrame({
    'tickers': ['TATA', 'WMT', 'MSFT'],
    'price': [845, 65, 64 ],
    'pe': [30.37, 14.26, 30.97],
    'eps': [27.82, 4.61, 2.12]
})

df_weather =  pd.DataFrame({
    'day': ['1/1/2019','1/2/2019','1/3/2019'],
    'temperature': [32,35,28],
    'event': ['Rain', 'Sunny', 'Snow']
})
```

#### use ExcelWriter Class;

```python
with pd.ExcelWriter('new_stock_weather.xlsx') as writer:
    df_stocks.to_excel(writer, sheet_name='Stocks')
    df_weather.to_excel(writer, sheet_name='Weather')
```
