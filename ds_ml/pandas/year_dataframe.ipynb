```python
odd_even = []
def add_odd_even_column(df):    
    for y in df["year"]:
        if y%2 == 0:
            odd_even.append("even")
        else:
            odd_even.append("odd")
    odd_even_series = pd.Series(odd_even)
    return df.assign(odd_even = odd_even_series)

df_year_odd_even = add_odd_even_column(df_year)
df_year_odd_even
```

```python
leap_year = []
def check_leap_year(df):
    for y in df["year"]:
        if ((y%4 == 0) & (y%100 != 0) | (y%400 == 0)):
            leap_year.append("yes")
        else: 
            leap_year.append("no")
    
    return df.assign(leap_year = pd.DataFrame(leap_year))

df_year_leap_odd_even = check_leap_year(df_year_odd_even)
df_year_leap_odd_even
```

```python
square_result = []
def perfect_square_year(df):
    for y in df["year"]:
        for i in range(y//2 + 1):
            if y == i*i:
                square_result.append(i)
                break
        else:
            square_result.append("no")
    return square_result
                
            
        
square_result_list = perfect_square_year(df_year_leap_odd_even)
df_year_all = df_year_leap_odd_even.assign(perfect_square = square_result_list)
df_year_all
```
