```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [50, 52, 51, 49, 48, 50, 47]
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y, 'g+')
```

```python
plt.plot(x,y, 'g--')
```

```python
plt.plot(x,y, 'D--r')
```

```python
plt.plot(x,y, 'd--r')
```

```python
plt.plot(x,y, 'd--b')
```

```python
plt.plot(x,y,color='red',marker='D',linestyle='',alpha=0.2)
```

```python
plt.plot(x,y,color='red',marker='D',linestyle='--',alpha=0.5)
```
