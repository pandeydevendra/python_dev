```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
expence_values = [1400, 600, 300, 410, 250]
expence_labels = ["Home Rent", "Food", "Phone/Internet Bill", "Car", "Other Utilities"] 
```

```python
plt.pie(expence_values, labels=expence_labels)
```

```python
plt.pie(expence_values, labels=expence_labels)
plt.show()
```

```python
plt.axis("equal")
plt.pie(expence_values, labels=expence_labels, radius=1.5, autopct='%0.2f%%')
plt.show()
```

```python
plt.axis("equal")
plt.pie(expence_values, labels=expence_labels, radius=1.5, autopct='%0.2f%%', shadow=True, explode=(0,0,0,0,0))
plt.show()
```

```python
plt.axis("equal")
plt.pie(expence_values, labels=expence_labels, radius=1.5, autopct='%0.2f%%', shadow=True,
        explode=(0.1,0.2,0.1,0.1,0.1))
plt.show()

plt.savefig("Expense_Pie_Chart.png", bbox_inches="tight", transparent=True)
```

```python
plt.axis("equal")
plt.pie(expence_values, labels=expence_labels, radius=1.5, autopct='%0.2f%%', shadow=True,
        explode=(0.05,0.2,0.12,0.1,0.15), startangle=0, counterclock=False)

plt.savefig("/home/vins/Desktop/Expense_Pie_Chart.png", bbox_inches="tight", pad_inches=2, transparent=False)

plt.show()
```
