```python
import requests
from bs4 import BeautifulSoup as soup
import pandas as pd
```

```python
# Define header:
header = {'Origin': 'https://www.1mg.com',
          'Referer': 'https://www.1mg.com/categories/exclusive/immunity-boosters/vitamin-c-734',
          'User-Agent': 'new_web_user'}
```

```python
# Send a get request:
url = 'https://www.1mg.com/categories/exclusive/immunity-boosters/vitamin-c-734'
html = requests.get(url)
bsobj = soup(html.content, 'html.parser')
```

```python
html.status_code
```

```python
bsobj
```

```python
# Get product name:
product_name = []
for name in bsobj.findAll('div', {'itemprop': 'name'}):
    print(name)
    product_name.append(name.text.strip())
```

```python
print(product_name)
```

```python
len(product_name)
```

```python
pack_size = []
for size in bsobj.findAll('div', {'class': 'style-pack-size-23QG7'}):
    print(size)
    pack_size.append(size.text.strip())
```

```python
print(pack_size)
```

```python
mrp = []
for price in bsobj.findAll('div', {'class': 'style-pack.tag-cOxYc'}):
    print(price)
    mrp.append(price.text.replace('$', '').replace('MRP', '').strip())
```

```python
print(mrp)
```

```python
# create pandas dataframe
# import pandas as pd

d1 = {'pname':product_name, 'psize': pack_size, 'mrp':mrp}
```
